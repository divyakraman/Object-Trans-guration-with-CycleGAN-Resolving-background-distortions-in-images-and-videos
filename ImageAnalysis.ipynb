```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
a = plt.imread('Results/Object Detection/failure1_fake_B.png')
b = plt.imread('Results/Object Detection/failure1_real_A.png')
plt.imshow(b)
```

```python
plt.figure(figsize=(10,2))
plt.subplot(1,3,1)
plt.title('Original Image')
plt.imshow(b)
plt.axis("off")
plt.subplot(1,3,2)
plt.title('CycleGAN Result')
plt.imshow(a)
plt.axis("off")
d = a
#Horse at d[~(80:240),~(35:225),:] 
d[:80,:,:]=b[:80,:,:]
d[241:,:,:]=b[241:,:,:]
d[80:240,0:40,:]=b[80:240,0:40,:]
d[80:240,240:,:]=b[80:240,240:,:]
plt.subplot(1,3,3)
plt.title('Our Result')
plt.imshow(d)
plt.axis("off")
```
